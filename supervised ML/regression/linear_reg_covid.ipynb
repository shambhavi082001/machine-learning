```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
# importing main dataset
df = pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
# keeping only required columns
df = df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
#rendering column namess
df.columns = ['date','state','cured','death','confirmed']
```

```python
df.head()
```

```python
df.tail()
```

```python
today = df[df.date == '2021-08-11']
```

```python
today.head()
```

```python
# sorting data with respect to number of confirmed cases
max_confirmed_cases = today.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
top_states_confirmed=max_confirmed_cases[0:5]
```

```python
top_states_confirmed
```

```python
#making bar plot for states with top confirmed cases
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='state',y='confirmed',data=top_states_confirmed,hue='state')
plt.show()
```

```python
## sorting data with respect to number of death cases
max_death_cases=today.sort_values(by='death',ascending=False)
max_death_cases
```

```python
max_death_cases.head()
```

```python
#states with maximum number of death cases
top_states_death=max_death_cases[0:5]
```

```python
# bar plot for states with top death case
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='state',y='death',data=top_states_death,hue='death')
plt.show()
```

```python
### sorting data with respect to number of cured cases
max_cured_cases=today.sort_values(by='cured',ascending=False)
max_cured_cases
```

```python
max_cured_cases.head()
```

```python
##states with maximum number of cured cases
top_states_cured=max_cured_cases[0:5]
```

```python
top_states_cured.head()
```

```python
# bar plot for states with top cured cases
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='state',y='cured',data=top_states_cured,hue='cured')
plt.show()
```
